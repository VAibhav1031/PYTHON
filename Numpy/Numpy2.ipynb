# COPY VS VIEW

```python
import  numpy as np
```

```python
# copy - copy  owns the data
# view- view doesn't own  the  data

# view - any  changes to array  will show changes in the view  also
# copy - any changes in the  original  will not  show  changes in the view
```

```python
arr=np.array([4,7,8,5])
```

```python
x=arr.view()
```

```python
arr[1]=5
```

```python
print(x)
```

```python
print(arr)
```

```python
y=arr.copy()
```

```python
print(y)
```

```python
y[2]=74
```

```python
y
```

```python
arr
```

**so till now  we see there is  no  change  occurs in the original  array   while  changes in original array**


## How  to check if array owns its data?

### *we  know that copies  owns  data*
### And view  does not own the data

### but how to  check this?

### Every numpy array  has attribute  base that returns None if the array owns the data
### otherwise the base attribute return the  original object

```python
print(x.base)
```

```python
print(y.base)
```
